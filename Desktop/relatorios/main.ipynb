```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

df = pd.read_csv("curva.csv")
df

```python
df["lucro"] = df.pecas / df.inteira
df
```

```python
plt.plot(df.aprov, df.lucro)
plt.grid()
```

```python
coef = np.polyfit(df.aprov, df.lucro, 1)
x = np.linspace(0.55, 0.8, 100)
y = np.polyval(coef, x)
```

```python
plt.figure()
plt.plot(x, y)
```

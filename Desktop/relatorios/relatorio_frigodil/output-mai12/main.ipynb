```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Caminho para a pasta com as imagens
pasta_imagens = "./output"


# Definir número de linhas e colunas da grade
linhas = 10
colunas = 5

# Tamanho desejado de cada imagem (ajustável conforme necessário)
tam_img = (100, 100)

# Lista de imagens na pasta
imagens = [
    img
    for img in os.listdir(pasta_imagens)
    if img.lower().endswith((".png", ".jpg", ".jpeg"))
]
imagens.sort()  # Opcional

# Criar figura com subplots
fig, axs = plt.subplots(linhas, colunas, figsize=(colunas * 1.5, linhas * 1.5))
plt.subplots_adjust(wspace=0.01, hspace=0.01)

# Desligar eixos e plotar as imagens
for i in range(linhas * colunas):
    row = i // colunas
    col = i % colunas
    ax = axs[row, col]
    ax.axis("off")

    if i < len(imagens):
        caminho = os.path.join(pasta_imagens, imagens[i])
        imagem = Image.open(caminho)
        # imagem = Image.open(caminho).resize(tam_img)
        ax.imshow(imagem)
        ax.text(
            5,
            15,
            f"{i+1}",
            fontsize=10,
            color="white",
            weight="bold",
            bbox=dict(facecolor="black", alpha=0.5, boxstyle="round,pad=0.2"),
        )

plt.tight_layout(pad=0.01)
plt.savefig("imagem.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
```

```python
df = pd.read_csv("dados_refinados.csv")
df.describe()
```

```python
df.PESO.hist(bins=10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset exemplo do Seaborn

# Criar o boxplot
# sns.boxplot(x="MATURIDADE", y="Score", data=df)
sns.boxplot(y="Score", data=df)

# Exibir o gráfico
plt.title("Boxplot do Score")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Filtrar apenas maturidade 2 e 4
df_filtrado = df[df["MATURIDADE"].isin([2, 4])]

# Gerar a tabela crosstab com média do Score
tabela = pd.crosstab(
    index=df_filtrado["ACABAMENTO"],
    columns=df_filtrado["MATURIDADE"],
    values=df_filtrado["Score"],
    aggfunc="mean",
)

print(tabela)

# Criar o heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(tabela, annot=True, cmap="viridis", fmt=".2f")

# Título e eixos
plt.title("Média do Score por ACABAMENTO (MATURIDADE 2 e 4)")
plt.xlabel("MATURIDADE")
plt.ylabel("ACABAMENTO")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

# Suponha que df já está definido e contém a coluna 'PESO'

# Dados
pesos = df['PESO'].dropna()  # remover valores nulos se houver

# Parâmetros da normal
media = pesos.mean()
desvio = pesos.std()

# Histograma
plt.figure(figsize=(8,5))
contagem, bins, _ = plt.hist(pesos, bins=25, color='skyblue', edgecolor='black', density=True)

# Curva normal
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, media, desvio)

# Plot da curva
plt.plot(x, p, 'r--', linewidth=2)
plt.title('Histograma do Peso com Curva Normal')
plt.xlabel('Peso')
plt.ylabel('Densidade')
plt.grid(True)
plt.show()
```
